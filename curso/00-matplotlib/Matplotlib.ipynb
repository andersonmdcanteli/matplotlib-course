## Introdução

O [matplotlib](https://matplotlib.org/stable/index.html) é uma biblioteca abrangente para a criação de visualizações *estáticas*, *animadas* e até *interativas* em **Python**. utilizando esta biblioteca, é possível:

- Criação:
    
    * Desenvolver gráficos de qualidade de publicação com apenas algumas linhas de código;

    * Utilizar figras interativas (zoom, pan, update);

- Customização

    * Fornece controle total aos estilos, propriedads das fontes, dos eixos, etc;
    
    * Exportação em diversos formatos;
    
- Extensivo:

    * Personalizado com diversos pacotes de terceiros;
    
Esta biblioteca fornece um interface similar a interface de geração de gráficos disponível no MATLAB, com a vantagem de ser ***gratuita e open-source***.

## Interfaces

Temos dois métodos para desenhar gráficos utilizando o matplotlib: utilizando a orientação a objetos ou de forma direta

Mas em ambos os casos, é necessário importar a biblioteca gráfica:

````python
import matplotlib.pyplot as plt
````

É dentro da classe ```matplotlib.pyplot``` que temos acesso a todas as funções para desenhar os gráficos. Como este termo é muito longo, é mais adequado dar um *alias* (apelido) à esta classe, que é o ```plt```.

<a id="estilo-matlab-id"></a>

### Forma direta (MATLAB style)

Esta forma utiliza as funções de forma bastante direta, que cria e gerencia as figuras e eixos de forma automática.

Exemplo:

````python
plt.plot([1,2,3,4,5],[1,2,3,5,7], label='Números primos') # criando o gráfico
plt.xlabel('x label') # adicionando nome no eixo x
plt.ylabel('y label') # adicionando nome no eixo y
plt.title('Meu título incrível') # adicionando titulo ao gráfico
plt.legend() # adicionando a legenda
plt.show() # apresentando o gráfico
````

<a id="o-o-id"></a>

### Orientação a objetos

Esta forma faz uso da orientação a objetos, que explicitamente cria figuras e eixos, adicionando métodos neles.

Exemplo:

````python
fig, ax = plt.subplots()  # criando uma instancia de figura e de eixo
ax.plot([1,2,3,4,5],[1,2,3,5,7], label='Números primos') # criando o gráfico
ax.set_xlabel('x label') # adicionando nome no eixo x
ax.set_ylabel('y label') # adicionando nome no eixo y
ax.set_title("Meu título incrível") # adicionando titulo ao gráfico
ax.legend() # adicionando a legenda
plt.show() # apresentando o gráfico
````

Nós iremos utilizar a primeira forma em toda a base do curso, de forma que você compreenda o funcionamento do **matplotlib**, pois esta é uma forma mais simples. Utilizar a orientação a objetos é mais eficiente, entretanto, requer maior conhecimento em Python, e por isso, vamos utilizar esta forma apenas quando necessário.

<a id="figure-id"></a>

### Figure

Este elemento corresponde a figura como um todo. Ela controla todos os demais elementos, como os *Axes*, *Axis* e *Artists*. Entenda este elemento como sendo a tela (em branco) de uma pintura .


<a id="axes-id"></a>

### Axes

O ```Axes``` corresponde a região da imagem com os dados preenchidos (a pintura). Cada figura pode conter diversos ```Axes```. Cada ```Axes``` pode conter dois (ou 3, no caso de um gráfico em 3 dimensões) ```Axis```, que é o elemento que determina os limites dos dados do gráfico. A ```Axes``` é a entrada primária para utilizar a orientação a objetos, e por isso vamos utiliza-la pouco.



<a id="axis-id"></a>

### Axis

Os ```Axis``` determinam os limites dos gráficos, os ticks e os tickslabels dos eixos.

<a id="artist-id"></a>

### Artist

Um ```Artist``` é basicamente tudo que você ve em um gráfico (inclusive os *Axis*, *Axes* e *Figure*). Então os objetos de texto, linhas, coleções, etc, são todos da classe ```Artist```. Quando a figura é renderizada, todos os artistas são desenhados no canvas.

## Key-points

- ```Figure``` é a imagem final que contem 1 ou mais ```Axes```;

- ```Axes``` representam um plot individual;

- Não confunda ```Axes``` com ```Axis``` (axis se refere aos eixos *x* e *y* (e *z*, se for um gráfico em 3D) do plot).
