## All avaiable arrows options

According to the [documentation](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.annotate.html), the available styles are as follows:

accessed on 02/05/2021

````python
arrow_styles = ['-', '->', '-[', '|-|', '-|>', '<-', '<->', '<|-', '<|-|>', 'fancy', 'simple', 'wedge']
````

## Generating data

We don't need data for this plot.

## Generating the plot

To plot the graph, I will create a figure with size of 8 by 8:

```python
plt.figure(figsize=(8,8))
```

So I'm going to create a for loop to iterate through all the elements contained in x ```arrow_styles```

```python
for i in range(len(arrow_styles)):
```

Now I’m going to add the annotated arrows.

The first parameter is the text to be annotated, which is a string composed of the symbol contained in ```arrow_styles\[i\]``` in single quotes:

```python
text = " '" + arrow_styles[i] + "' ", 
```

The second is the coordinate where the annotate element starts. It will always start at ```x = 1```, but the value of ```y``` will vary according to the value ```i```, so that each annotation is one unit above the previous one.

```python
xy = (1, i), 
```

The third is the coordinate where the text will be inserted. It will always be at ```x = 6```, but the value of ```y``` will vary with the value of ```i```, causing the arrow line to be parallel to the *x axis*.

```python
xytext = (6, i), 
```

The ```va``` parameter as ```'center'``` is necessary to ensure that the arrow is parallel to the *x axis*.

```python
va='center'
```

The ```fontsize``` parameter is to slightly increase the font size used.

```python
fontsize = 14,
```

In the ```arrowprops``` parameter is where we effectively change the type of the arrowhead. This parameter receives a dictionary with the properties of the arrow.

```python
arrowprops={
     ...
}
```

For the ```arrowstyle``` key I pass the symbol used to change the arrow, which is contained in list ```arrow_styles``` in position ```i```.


```python
'arrowstyle': arrow_styles[i], 
```

The ```lw``` key is used to change the thickness of the arrow line.

```python
'lw': 1.5, 
```

The ```facecolor``` parameter changes the color of the arrow to white.

```python
'facecolor':'w', 
```

The ```edgecolor``` parameter changes the border color to black.

```python
'edgecolor': 'k'
```

The ```plt.xlim()``` element sets the graph's x-axis range.

```python
plt.xlim(0.2,7.5)
```

The ```plt.ylim()``` element sets the graph's y-axis range.

```python
plt.ylim(-1,12)
```

The element ```plt.xticks()``` changes the ticks of the x axis. The empty list removes them.

```python
plt.xticks([])
```

The element ```plt.yticks()``` changes the ticks of the y axis. The empty list removes them.

```python
plt.yticks([])
```

The ```plt.title()```add a title to the plot. The ```pad``` adds a pading to the title.

```python
plt.title("Arrow styles for plt.annotate", pad=15)
```

The ```plt.savefig()``` create a file from the figure. The ```dpi``` parameter sets the number of pixels per inches, and the ```bbox_inches='tight'``` removes extra margins from the chart

```python
plt.savefig("arrow_styles.png", dpi=300, bbox_inches='tight')
```


````python
plt.figure(figsize=(8,8))
for i in range(len(arrow_styles)):
    plt.annotate(text = " '" + arrow_styles[i] + "' ", 
                 xy = (1, i), 
                 xytext = (6, i), 
                 va='center',
                 fontsize = 14, 
                 arrowprops={'arrowstyle': arrow_styles[i], 
                             'lw': 1.5, 
                             'facecolor':'w', 
                             'edgecolor': 'k'}
                )
plt.xlim(0.2,7.5)
plt.ylim(-1,12)
plt.xticks([])
plt.yticks([])
plt.title("Arrow styles for plt.annotate", pad=15)
plt.savefig("arrow_styles.png", dpi=300, bbox_inches='tight')
plt.show()    
````
