```python
import ternary
import matplotlib.pyplot as plt
```

```python
flavors = ['Lemon', 'Chocolate', 'Strawberry']
```

```python
# Creating plot with oo style 
fig, tax = ternary.figure()

# setting the figure size
fig.set_size_inches(8, 8)

# setting the background color of the figure
fig.set_facecolor("white")

# Removing the axes created by matplotlib
tax.get_axes().axis('off')
tax.clear_matplotlib_ticks()

# Setting the graphic background color
tax.set_background_color('white')

# Inserting ticks
tax.ticks(axis='lbr', multiple=0.2, tick_formats="%.2f", linewidth=1, offset=0.02, 
          axes_colors={'l': '#FF1493',
                       'r': '#D2691E',
                       'b': '#228B22'})

# Inserting limits in the graph (the axes) 
tax.boundary(axes_colors={'l': '#FF1493',
                          'r': '#D2691E', 
                          'b': '#228B22'})
# Naming the bottom axis
tax.bottom_axis_label(flavors[0], offset=0.1, color='#228B22', fontsize=16)
# or
tax.right_corner_label(flavors[0], offset=-.05, color='#228B22', fontsize=16)

# Naming the right axis 
tax.top_corner_label(flavors[1], offset = 0.2, color='#D2691E', fontsize=16)
# or
tax.right_axis_label(flavors[1], offset = 0.15, color='#D2691E', fontsize=16)

# Naming the left axis
tax.left_corner_label(flavors[2], offset=0.4, color='#FF1493', fontsize=16)
# or
tax.left_axis_label(flavors[2], offset=0.15, color='#FF1493', fontsize=16)

plt.arrow(x=0.3, y=0.1, dx=.45, dy=.0, width=.01, color='#228B22', length_includes_head=True)

plt.arrow(x=0.75, y=0.2, dx=-.2, dy=.35, width=.01, color='#D2691E', length_includes_head=True)

plt.arrow(x=0.45, y=0.55, dx=-.2, dy=-.35, width=.01, color='#FF1493', length_includes_head=True)

# exporting
tax.savefig("axis-summary.png", bbox_inches='tight', dpi=300)

# showing the figure
tax.show()
```
